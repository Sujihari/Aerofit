# Data Description:
**Product Purchased:** Categorical variable indicating the type of treadmill purchased, with options KP281, KP481, or KP781.

**Age:** Numerical variable representing the age of the customers in years.

**Gender:** Categorical variable indicating the gender of the customers, with options Male or Female.

**Education:** Numerical variable representing the number of years of education completed by the customers.

**MaritalStatus:** Categorical variable indicating the marital status of the customers, with options Single or partnered.

**Usage:** Numerical variable indicating the average number of times the customer plans to use the treadmill each week.

**Income:** Numerical variable representing the annual income of the customers in dollars.

**Fitness:** Numerical variable representing the self-rated fitness level of the customers on a scale of 1 to 5, where 1 is poor 
shape and 5 is excellent shape.

**Miles:** Numerical variable indicating the average number of miles the customer expects to walk/run each week.

```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans

# Step 3: Feature Engineering (creating age groups and income brackets)
# Example: Categorizing age into age groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 40, 50, 60, float('inf')],
                        labels=['<30', '30-40', '40-50', '50-60', '60+'])

# Example: Categorizing income into income brackets
df['IncomeBracket'] = pd.cut(df['Income'], bins=[0, 30000, 50000, 70000, 100000, float('inf')],
                             labels=['<30K', '30K-50K', '50K-70K', '70K-100K', '100K+'])

# Step 4: Select relevant features for clustering
features = ['AgeGroup', 'Gender', 'MaritalStatus', 'IncomeBracket']

# Step 5: One-hot encoding categorical variables
df_encoded = pd.get_dummies(df[features])

# Step 6: Perform clustering using K-means
kmeans = KMeans(n_clusters=5, random_state=42)
df['Segment'] = kmeans.fit_predict(df_encoded)

# Step 7: Interpretation and Analysis
# Analyze characteristics of each segment, visualize distributions, compute summary statistics, etc.

# Step 8: Naming and Describing Segments
# Give meaningful names and descriptions to each segment based on their characteristics

# Step 9: Validation and Iteration
# Validate the identified segments to ensure they are meaningful and actionable
```
